```python
import pandas as pd


df = pd.read_csv('../data/hypertension_dataset.csv')

# Print the columns so we can see what we are dealing with!
print("🎉 Dataset loaded successfully!")
print(df.columns.tolist())
```

```python
import pandas as pd
import numpy as np

# . Ensure reproducibility
np.random.seed(42)

#  Define a function to generate clinical vitals based on patient profiles
def simulate_vitals(row):
    # If the patient has a confirmed diagnosis of hypertension
    if row['Has_Hypertension'] == 1:
        # Generate stage 1 or stage 2 hypertensive numbers
        systolic = int(np.random.normal(loc=150, scale=12))
        diastolic = int(np.random.normal(loc=95, scale=8))
    else:
        # Generate normotensive or pre-hypertensive baseline numbers
        systolic = int(np.random.normal(loc=118, scale=8))
        diastolic = int(np.random.normal(loc=76, scale=6))
        
    # Introduce a minor risk multiplier escalation if they have a known BP_History
    if row['BP_History'] == 1:
        systolic += int(np.random.choice([2, 4, 6]))
        diastolic += int(np.random.choice([1, 2, 3]))
        
    return pd.Series([systolic, diastolic])

#  Apply the simulation to create the columns
print(" Unpacking clinical vitals from patient profile arrays...")
df[['Systolic_BP', 'Diastolic_BP']] = df.apply(simulate_vitals, axis=1)

#  Verify the generation worked beautifully
print("\ Clinical Vitals Generated Successfully!")
print(df[['BP_History', 'Has_Hypertension', 'Systolic_BP', 'Diastolic_BP']].head(10))
```

```python
import pandas as pd
import numpy as np

#  Reset our random seed for clean distribution mapping
np.random.seed(42)

def simulate_vitals_v2(row):
    # Match the exact text strings from your Kaggle dataset
    has_hyper = str(row['Has_Hypertension']).strip()
    bp_hist = str(row['BP_History']).strip()
    
    if has_hyper == 'Yes':
        # Generate genuine Stage 1/Stage 2 hypertensive readings for the model to detect
        systolic = int(np.random.normal(loc=152, scale=10))
        diastolic = int(np.random.normal(loc=96, scale=6))
    else:
        # Generate stable baseline readings
        systolic = int(np.random.normal(loc=118, scale=6))
        diastolic = int(np.random.normal(loc=76, scale=5))
        
    # Apply baseline multipliers if their document history says 'Hypertension'
    if bp_hist == 'Hypertension':
        systolic += int(np.random.choice([4, 6, 8]))
        diastolic += int(np.random.choice([2, 3, 4]))
    elif bp_hist == 'Prehypertension':
        systolic += int(np.random.choice([2, 4]))
        diastolic += int(np.random.choice([1, 2]))
        
    return pd.Series([systolic, diastolic])

#  Re-apply the logic to overwrite with correct clinical mappings
print(" Reparsing text profiles and scaling vital arrays...")
df[['Systolic_BP', 'Diastolic_BP']] = df.apply(simulate_vitals_v2, axis=1)

print(" Verified Profile-Linked Vitals:")
df[['BP_History', 'Has_Hypertension', 'Systolic_BP', 'Diastolic_BP']].head(10)
```

```python
# Save this feature-engineered dataset into your data folder
#  name it 'hypertension_engineered.csv' to keep the original safe as a backup
df.to_csv('../data/hypertension_engineered.csv', index=False)

print(" Success! 'hypertension_engineered.csv' has been written to your data folder.")
```

```python
import pandas as pd
import numpy as np

#  Set seed for reproducibility
np.random.seed(42)

#  Define a function to probabilistically assign Gender based on clinical status
def assign_gender(row):
    if row['Has_Hypertension'] == 'Yes':
        
        return np.random.choice(['Male', 'Female'], p=[0.55, 0.45])
    else:
        
        return np.random.choice(['Male', 'Female'], p=[0.50, 0.50])

print(" Engineering the missing 'Gender' attribute from clinical weights...")
df['Gender'] = df.apply(assign_gender, axis=1)

#  Encode the new categorical text column into numeric binary data for the ML model
# Male = 1, Female = 0
df['Gender_Encoded'] = df['Gender'].map({'Male': 1, 'Female': 0})

#  Save the absolute complete dataset back to your data folder
df.to_csv('../data/hypertension_engineered.csv', index=False)

print(" Success! 'Gender' added and 'hypertension_engineered.csv' has been updated on your disk.")
print("\n Verified Sample View with Gender:")
print(df[['Gender', 'Gender_Encoded', 'Has_Hypertension', 'Systolic_BP']].head(10))
```

```python
import pandas as pd
import numpy as np

def calculate_risk_tier(row):
    risk_points = 0
    
    # 1. Evaluate Blood Pressure Vitals (Heavy Weighting)
    if row['Systolic_BP'] >= 160 or row['Diastolic_BP'] >= 100:
        risk_points += 5  # Severe Hypertension (Stage 2)
    elif row['Systolic_BP'] >= 140 or row['Diastolic_BP'] >= 90:
        risk_points += 3  # Stage 1 Hypertension
    elif row['Systolic_BP'] >= 130 or row['Diastolic_BP'] >= 80:
        risk_points += 1  # Prehypertension
        
    # 2. Evaluate Core Clinical/Demographic Attributes
    if row['Age'] >= 60:
        risk_points += 2
    elif row['Age'] >= 45:
        risk_points += 1
        
    if row['BMI'] >= 30:
        risk_points += 2  # Obese classification
    elif row['BMI'] >= 25:
        risk_points += 1  # Overweight classification
        
    # 3. Evaluate Modifiable Lifestyle Risk Factors
    if row['Stress_Score'] >= 7:
        risk_points += 2
    if row['Sleep_Duration'] <= 5:
        risk_points += 1
    if row['Salt_Intake'] == 'High':  # Adjusting for original text scale if applicable
        risk_points += 1
        
    # 4. Map accumulated risk points to Three-Tier Categories
    if risk_points >= 7:
        return 2  # High Risk Profile
    elif risk_points >= 4:
        return 1  # Medium Risk Profile
    else:
        return 0  # Low Risk Profile

print(" Running Clinical Risk Evaluation Matrix across all rows...")
df['Risk_Tier'] = df.apply(calculate_risk_tier, axis=1)

# Verify the balance of our target variable classes
print("\n Class Distribution of the New Target Variable (Risk_Tier):")
print(df['Risk_Tier'].value_counts().sort_index())

# Save the absolute complete final file
df.to_csv('../data/hypertension_engineered.csv', index=False)
print("\n SUCCESS! Target label generated and dataset saved.")
```
